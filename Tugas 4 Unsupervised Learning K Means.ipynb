```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
mallCustomersDf = pd.read_csv("https://raw.githubusercontent.com/pavankalyan1997/Machine-learning-without-any-libraries/master/2.Clustering/1.K_Means_Clustering/Mall_Customers.csv")
mallCustomersDf
```

```python
def euclidean(array1,array2,b1,b2):
    arrayjarak = []
    for i in range(len(array1)):
        jarak = np.round(((array1[i]-b1)**2)+((array2[i]-b2)**2))
        arrayjarak.append(jarak)
    return arrayjarak
```

```python
def knearest(array1,array2,array3,b1,b2):
    traindata = []
    arrayjarak = euclidean(array1,array2,b1,b2)
    traindata.append(array1)
    traindata.append(array2)
    traindata.append(array3)
    traindata.append(arrayjarak)
    return traindata
```

```python
mallCustomersDf.head(10)
```

```python
genre = mallCustomersDf['Genre'].values
genre
```

```python
income = mallCustomersDf['Annual Income (k$)'].values
income
```

```python
spending = mallCustomersDf['Spending Score (1-100)'].values
spending
```

```python
traindata = knearest(income,spending,genre,18,94)
df = pd.DataFrame({"Annual Income":traindata[0], 
                   "Spending Score(1-100)":traindata[1],
                   "Genre":traindata[2],
                   "Jarak ke data uji (X1=17,X2=25)":traindata[3]})
df.head(10)
```

```python
result = df.head(10).sort_values(by=['Jarak ke data uji (X1=17,X2=25)'])
result
```
